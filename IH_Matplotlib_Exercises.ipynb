## Exercise 1

* **Create a figure object called fig using plt.figure() **
* **Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* **Plot (x,y) on that axes and set the labels and titles to match the plot below: **

```python
# create a figure object
fig = plt.figure()

# add axes to the figure canvas at [0, 0, 1, 1]
axes = fig.add_axes([0, 0, 1, 1,])

# plot (x, y)
axes.plot(x, y, 'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
# create a figure object and put two axes on it ax1 and ax2
fig = plt.figure()

ax1 = fig.add_axes([0.0, 0.0, 1.0, 1.0]) # main axes

ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2]) # inset axes
```

**Now plot (x,y) on both axes. And call your figure object to show it.**

```python
fig = plt.figure()

ax1 = fig.add_axes([0.0, 0.0, 1.0, 1.0]) # main axes
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2]) # inset axes

ax1.plot(x, y, 'r') # r is red
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('object1')

ax2.plot(x, y, 'r') 
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('object2')
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig3 = plt.figure()

axes_1 = fig3.add_axes([0, 0, 1, 1]) # main axes
axes_2 = fig3.add_axes([0.2, 0.5, 0.4, 0.4]) # inset axes

# Larger Figure Axes 1
axes_1.plot(x, z, 'b')
axes_1.set_xlabel('X')
axes_1.set_ylabel('Z')

# Insert Figure Axes 2
axes_2.plot(x, y, 'b') 
axes_2.set_xlabel('X')
axes_2.set_ylabel('Y')
axes_2.set_title('zoom');

# set x limits and y limits
axes_2.set_ylim(ymin=30, ymax=50)
axes_2.set_xlim(xmin=20.0, xmax=22.0)
```
